### Question 1.1 (Instacart)
The UX design team wants to optimize the digital shelf layout. To determine if cart addition sequence can guide layout priority, does the order in which items are added to the shopping cart predict their reorder rate? Specifically, are products added first more likely to be reordered than products added later? Use only the prior order history.

````python
import pandas as pd
import numpy as np
import kagglehub, os

instacart_path = kagglehub.dataset_download("psparks/instacart-market-basket-analysis")
df_prior = pd.read_csv(os.path.join(instacart_path, "order_products__prior.csv"))

# Pearson correlation between cart addition order and reorder flag
corr = df_prior['add_to_cart_order'].corr(df_prior['reordered'])
print(f"Pearson correlation (add_to_cart_order vs reordered): {corr:.4f}")

# Mean reorder rate by cart position (first 15 positions)
pos_reorder = (
    df_prior.groupby('add_to_cart_order')['reordered']
    .mean()
    .reset_index()
    .rename(columns={'add_to_cart_order': 'cart_position', 'reordered': 'reorder_rate'})
)
print("\nMean reorder rate by cart position (positions 1–15):")
print(pos_reorder.head(15).to_string(index=False))
````

### Conclusion – Q1.1

**Yes, cart addition sequence is a statistically meaningful (though weak) predictor of reorder rate.**

- Items added **first** (position 1) have a reorder rate of **~67.8 %**, which declines monotonically to **~55 %** by position 10 and continues falling thereafter.
- The Pearson correlation between `add_to_cart_order` and `reordered` is **−0.133** — negative, as expected (earlier position → higher reorder rate).
- The effect is consistent and monotone, confirming a genuine behavioral pattern: shoppers tend to add habitual staples first, driving their higher reorder rate. Items added later are more exploratory or impulse-driven and are therefore reordered less frequently.

**Recommendation for UX:** Placing high-reorder-rate items near the top of the digital shelf (first scroll positions) may reinforce habitual purchase patterns and increase basket size.

### Question 1.2 (Instacart)
The customer segmentation team is analyzing user exploration patterns to design personalized promotions. To determine if users with broader shopping tastes are more loyal, do users who shop from a wider variety of unique aisles (high aisle diversity) have a higher reorder rate on average? Calculate the correlation between a user's unique aisle count and their overall reorder rate.

````python
import pandas as pd
import numpy as np
import kagglehub, os

instacart_path = kagglehub.dataset_download("psparks/instacart-market-basket-analysis")
df_prior    = pd.read_csv(os.path.join(instacart_path, "order_products__prior.csv"))
df_orders   = pd.read_csv(os.path.join(instacart_path, "orders.csv"))
df_products = pd.read_csv(os.path.join(instacart_path, "products.csv"))

# Attach user_id and aisle_id to each prior order-product row
prior_u = df_prior.merge(df_orders[['order_id','user_id']], on='order_id')
prior_u = prior_u.merge(df_products[['product_id','aisle_id']], on='product_id')

# Per-user: unique aisle count and mean reorder rate
user_stats = prior_u.groupby('user_id').agg(
    unique_aisle_count=('aisle_id', 'nunique'),
    reorder_rate=('reordered', 'mean')
).reset_index()

corr = user_stats['unique_aisle_count'].corr(user_stats['reorder_rate'])
print(f"Pearson correlation (unique_aisle_count vs reorder_rate): {corr:.4f}")
print(f"\nUser-level summary stats:")
print(user_stats[['unique_aisle_count','reorder_rate']].describe().round(4).to_string())
````

### Conclusion – Q1.2

**Yes, users who shop from a wider variety of aisles do have a higher reorder rate on average.**

- The Pearson correlation between a user's unique aisle count and their overall reorder rate is **+0.3141** — a moderate positive relationship.
- The average user shops **~27.8 unique aisles** and has a reorder rate of **~43.2 %**.
- The result is counter-intuitive at first glance: one might expect "exploratory" shoppers (many aisles) to reorder less because they try new things. Instead, broad-aisle shoppers appear to be higher-frequency, habitual users who simply have more staples across more categories.

**Recommendation for segmentation:** Users with high aisle diversity are strong candidates for loyalty/VIP programs — they are already engaged across the platform and have demonstrated habitual purchasing behavior.

3. Aggregated per user: `nunique('aisle_id')` for diversity, `mean('reordered')` for loyalty rate.
4. Computed Pearson correlation across 206,209 users.

#### Data & Methodology Choices
- **Unit of analysis:** User level (not order level) — the question asks about users' reorder rates.
- **Aisle diversity metric:** Count of unique aisles shopped, simple and interpretable.
- **Why Pearson?** Both metrics are continuous; 206 K observations makes the test very powerful. Checked distributional reasonableness via describe().
- **Alternative considered:** Spearman correlation (rank-based), which would be appropriate if the relationship were nonlinear. Given the moderate magnitude of Pearson r = 0.31, the linear approximation is adequate.
- **Potential confound:** High aisle count is also correlated with total order count; prolific shoppers have higher diversity by definition. A partial-correlation controlling for order count could sharpen the inference.

### Question 1.5 (Steam)
The publisher pricing strategy team is evaluating the relationship between pricing premium and product quality perception. To verify if a price premium signals higher satisfaction, do higher-priced games receive better user ratings (positive review ratio)? Calculate the correlation between price and positive review ratio.

````python
import pandas as pd
import numpy as np
import kagglehub, os

steam_path = kagglehub.dataset_download("nikdavis/steam-store-games")
df_steam = pd.read_csv(os.path.join(steam_path, "steam.csv"))

# Compute positive review ratio; keep only games with at least 1 rating
total_ratings = df_steam['positive_ratings'] + df_steam['negative_ratings']
df_s = df_steam[total_ratings > 0].copy()
df_s['pos_ratio'] = df_s['positive_ratings'] / total_ratings

# All games (including free)
corr_all = df_s['price'].corr(df_s['pos_ratio'])

# Paid games only (price > 0)
df_paid = df_s[df_s['price'] > 0]
corr_paid = df_paid['price'].corr(df_paid['pos_ratio'])

print(f"Pearson correlation (price vs pos_ratio) — all games:   {corr_all:.4f}")
print(f"Pearson correlation (price vs pos_ratio) — paid games:  {corr_paid:.4f}")
print(f"\nGames with ratings: {len(df_s):,}")
print(f"Paid games with ratings: {len(df_paid):,}")
print(f"\nPrice quartile breakdown (paid games):")
df_paid = df_paid.copy()
df_paid['price_quartile'] = pd.qcut(df_paid['price'], q=4, labels=['Q1 (low)','Q2','Q3','Q4 (high)'])
print(df_paid.groupby('price_quartile', observed=True)['pos_ratio'].mean().round(4).to_string())
````

### Conclusion – Q1.5

**Higher-priced games do NOT meaningfully receive better user ratings.**

- The Pearson correlation between price and positive review ratio is **+0.083** (paid games only) — nearly zero.
- Even when breaking paid games into price quartiles, the mean positive-review ratio is roughly flat across Q1–Q4 (~70–73%).
- Price premium signals **commercial positioning**, not satisfaction. Users rate games on gameplay quality, not price paid.

**Recommendation for pricing strategy:** Setting a higher price does not automatically improve perceived quality or review scores. Content quality, polish, and community support are stronger drivers of positive ratings than price alone.

### Question 2.1 (Steam)

````python
import pandas as pd
import numpy as np
import kagglehub, os
from itertools import combinations

steam_path = kagglehub.dataset_download("nikdavis/steam-store-games")
df_steam = pd.read_csv(os.path.join(steam_path, "steam.csv"))

# ── Part 2: Genre synergy matrix ────────────────────────────────────────────
df_g = df_steam[['appid','publisher','genres']].dropna(subset=['genres','publisher'])
df_g = df_g.assign(genre=df_g['genres'].str.split(';')).explode('genre')
df_g['genre'] = df_g['genre'].str.strip()

# For each publisher, collect set of genres published
pub_genres = df_g.groupby('publisher')['genre'].apply(set).reset_index()
pub_genres.columns = ['publisher','genre_set']

# Count co-occurrences of genre pairs in publisher portfolios
co_counts = {}
for genres in pub_genres['genre_set']:
    for g1, g2 in combinations(sorted(genres), 2):
        co_counts[(g1, g2)] = co_counts.get((g1, g2), 0) + 1

# Top synergy pairs
synergy_pairs = sorted(co_counts.items(), key=lambda x: -x[1])
print("Top 15 genre synergy pairs (co-occurrence in publisher portfolios):")
print(f"{'Genre A':<25} {'Genre B':<25} {'Co-occurrences':>15}")
print("-" * 65)
for (g1, g2), cnt in synergy_pairs[:15]:
    print(f"{g1:<25} {g2:<25} {cnt:>15}")

# Compact synergy matrix for top genres
top_genres = ['Action','Adventure','Casual','Indie','RPG','Simulation','Sports','Strategy']
synergy_df = pd.DataFrame(0, index=top_genres, columns=top_genres)
for (g1, g2), cnt in co_counts.items():
    if g1 in top_genres and g2 in top_genres:
        synergy_df.loc[g1, g2] = cnt
        synergy_df.loc[g2, g1] = cnt
print("\nGenre Synergy Matrix (publisher portfolio co-occurrences):")
print(synergy_df.to_string())
````

````python
import pandas as pd
import numpy as np
import kagglehub, os

instacart_path = kagglehub.dataset_download("psparks/instacart-market-basket-analysis")
df_prior       = pd.read_csv(os.path.join(instacart_path, "order_products__prior.csv"))
df_orders      = pd.read_csv(os.path.join(instacart_path, "orders.csv"))
df_products    = pd.read_csv(os.path.join(instacart_path, "products.csv"))
df_departments = pd.read_csv(os.path.join(instacart_path, "departments.csv"))

prior_orders = df_orders[df_orders['eval_set'] == 'prior']
user_order_count = prior_orders.groupby('user_id')['order_id'].count().reset_index(name='order_count')
top50 = user_order_count.nlargest(50, 'order_count')

# Get all prior order_ids for top-50 users
top50_order_ids = prior_orders[prior_orders['user_id'].isin(top50['user_id'])]['order_id']

# Attach user_id and department to each purchased product
top50_prior = (
    df_prior[df_prior['order_id'].isin(top50_order_ids)]
    .merge(df_orders[['order_id','user_id']], on='order_id')
    .merge(df_products[['product_id','department_id']], on='product_id')
    .merge(df_departments, on='department_id')
)

# Preferred department per user (most items from that department)
user_dept_counts = top50_prior.groupby(['user_id','department']).size().reset_index(name='dept_count')
user_pref_dept = user_dept_counts.loc[
    user_dept_counts.groupby('user_id')['dept_count'].idxmax()
][['user_id','department','dept_count']]

# Department focus: % of all items from the preferred department
user_total = top50_prior.groupby('user_id').size().reset_index(name='total_items')
result = user_pref_dept.merge(user_total, on='user_id')
result['dept_focus_pct'] = (result['dept_count'] / result['total_items'] * 100).round(2)
result = result.sort_values('user_id').reset_index(drop=True)

print("VIP Top-50 Users — Preferred Department & Focus Score:")
print(result[['user_id','department','dept_focus_pct']].to_string(index=False))
print(f"\nOverall mean department focus: {result['dept_focus_pct'].mean():.2f}%")
print(f"\nPreferred department distribution:")
print(result['department'].value_counts().to_string())
````

### Question 2.3 (Instacart)
To optimize warehouse turnaround, the logistics team wants to test if baskets anchored by 'essential' groceries predict higher platform reorder rates. Segment users into quartiles based on their consumption of essential items, and compare the overall reorder rate of the highest quartile (essential-heavy) against the lowest quartile (discretionary).

````python
import pandas as pd
import numpy as np
import kagglehub, os

instacart_path = kagglehub.dataset_download("psparks/instacart-market-basket-analysis")
df_prior       = pd.read_csv(os.path.join(instacart_path, "order_products__prior.csv"))
df_orders      = pd.read_csv(os.path.join(instacart_path, "orders.csv"))
df_products    = pd.read_csv(os.path.join(instacart_path, "products.csv"))
df_departments = pd.read_csv(os.path.join(instacart_path, "departments.csv"))

# Define essential departments: staple grocery categories
ESSENTIAL_DEPTS = {'produce', 'dairy eggs', 'meat seafood', 'bakery', 'beverages'}
essential_dept_ids = set(
    df_departments[df_departments['department'].isin(ESSENTIAL_DEPTS)]['department_id']
)
print(f"Essential departments: {ESSENTIAL_DEPTS}")
print(f"Mapped dept_ids: {essential_dept_ids}")

# Attach department and user_id to prior order products
prior_dept = df_prior.merge(df_products[['product_id','department_id']], on='product_id')
prior_dept['is_essential'] = prior_dept['department_id'].isin(essential_dept_ids).astype(int)
prior_dept_user = prior_dept.merge(df_orders[['order_id','user_id']], on='order_id')

# Per-user: essential consumption ratio and overall reorder rate
user_stats = prior_dept_user.groupby('user_id').agg(
    essential_ratio=('is_essential', 'mean'),
    reorder_rate=('reordered', 'mean')
).reset_index()

# Segment into quartiles by essential_ratio
user_stats['quartile'] = pd.qcut(user_stats['essential_ratio'], q=4,
                                  labels=['Q1 (Discretionary)','Q2','Q3','Q4 (Essential-heavy)'])

summary = user_stats.groupby('quartile', observed=True)['reorder_rate'].agg(
    mean_reorder_rate='mean', count='count'
).reset_index()
print("\nReorder rate by essential consumption quartile:")
print(summary.to_string(index=False))

q4 = user_stats[user_stats['quartile']=='Q4 (Essential-heavy)']['reorder_rate'].mean()
q1 = user_stats[user_stats['quartile']=='Q1 (Discretionary)']['reorder_rate'].mean()
print(f"\nQ4 (essential-heavy) mean reorder rate: {q4:.4f}")
print(f"Q1 (discretionary)   mean reorder rate: {q1:.4f}")
print(f"Difference (Q4 − Q1): {q4 - q1:.4f}  ({(q4 - q1)/q1*100:.1f}% relative lift)")
````

| Q2 | 41.62% |

- The Q4 vs Q1 difference is **+10.87 percentage points** (~28.8% relative lift).
- The monotone increase across all four quartiles confirms this is not a spurious result — more essential purchasing → higher reorder rate at every tier.

````python
import pandas as pd
import numpy as np
import kagglehub, os

instacart_path = kagglehub.dataset_download("psparks/instacart-market-basket-analysis")
df_orders = pd.read_csv(os.path.join(instacart_path, "orders.csv"))

# In the Instacart dataset, order_dow 0 = Saturday, 1 = Sunday
# Sort by user and order number to correctly identify "next order"
df_o = df_orders.sort_values(['user_id','order_number']).copy()

# Shift days_since_prior_order UP by 1 → each row gets the delay of the FOLLOWING order
df_o['delay_after_this_order'] = df_o.groupby('user_id')['days_since_prior_order'].shift(-1)

# Classify weekends (0=Sat, 1=Sun) vs weekdays (2–6)
df_o['is_weekend'] = df_o['order_dow'].isin([0, 1])

# Mean delay to next order grouped by order_dow
dow_delay = (
    df_o.dropna(subset=['delay_after_this_order'])
    .groupby('order_dow')['delay_after_this_order']
    .agg(mean_delay='mean', median_delay='median', count='count')
    .reset_index()
)
dow_labels = {0:'Sat', 1:'Sun', 2:'Mon', 3:'Tue', 4:'Wed', 5:'Thu', 6:'Fri'}
dow_delay['day_name'] = dow_delay['order_dow'].map(dow_labels)
print("Mean delay (days) to next order, by day order was placed:")
print(dow_delay[['day_name','mean_delay','median_delay','count']].to_string(index=False))

weekend = df_o[df_o['is_weekend'] & df_o['delay_after_this_order'].notna()]['delay_after_this_order']
weekday = df_o[~df_o['is_weekend'] & df_o['delay_after_this_order'].notna()]['delay_after_this_order']
print(f"\nWeekend (Sat/Sun) mean delay: {weekend.mean():.4f} days  |  median: {weekend.median():.1f} days")
print(f"Weekday (Mon-Fri) mean delay: {weekday.mean():.4f} days  |  median: {weekday.median():.1f} days")
print(f"Difference (Weekend − Weekday): {weekend.mean()-weekday.mean():.4f} days")
````

### Question 3.2 (Steam)
A branding agency is auditing marketing copy styles across developers to detect ghostwriting or studio style replication. Given that Valve has iconic language in its descriptions (e.g., Half-Life 2), can we identify other games developed by Valve using only the similarity of game descriptions?

````python
import pandas as pd
import numpy as np
import kagglehub, os
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

steam_path = kagglehub.dataset_download("nikdavis/steam-store-games")
df_steam = pd.read_csv(os.path.join(steam_path, "steam.csv"))
df_desc  = pd.read_csv(os.path.join(steam_path, "steam_description_data.csv"))

# Ground-truth Valve games (developer == 'Valve' exactly)
valve_appids = set(df_steam[df_steam['developer'] == 'Valve']['appid'])
print(f"Known pure-Valve games: {len(valve_appids)}")

# Prepare descriptions
df_d = df_desc[['steam_appid','detailed_description']].dropna(subset=['detailed_description'])
print(f"Games with descriptions: {len(df_d)}")

# TF-IDF over detailed descriptions
tfidf = TfidfVectorizer(max_features=5000, stop_words='english', min_df=2)
tfidf_matrix = tfidf.fit_transform(df_d['detailed_description'])

# Indices of known Valve games in the description matrix
valve_idx = np.where(df_d['steam_appid'].isin(valve_appids))[0]
print(f"Valve games in description corpus: {len(valve_idx)}")

# Mean TF-IDF vector of all known Valve games
valve_centroid = np.asarray(tfidf_matrix[valve_idx].mean(axis=0))

# Cosine similarity of all games vs. Valve centroid
sims = cosine_similarity(valve_centroid, tfidf_matrix)[0]
df_d = df_d.copy()
df_d['valve_sim'] = sims

# Top 20 non-Valve games most similar to Valve
non_valve_mask = ~df_d['steam_appid'].isin(valve_appids)
top20 = df_d[non_valve_mask].nlargest(20, 'valve_sim')
top20 = top20.merge(df_steam[['appid','name','developer']], left_on='steam_appid', right_on='appid', how='left')
print("\nTop 20 non-Valve games most similar to Valve description style:")
print(top20[['steam_appid','name','developer','valve_sim']].to_string(index=False))
````

### Conclusion – Q3.2

**Yes — description-based TF-IDF similarity can identify games that are stylistically and genre-wise close to Valve's portfolio**, though it cannot perfectly recover the developer label without ground truth.

**Top matches by description similarity to Valve games:**
1. ROM: Extraction (First Contact Entertainment) — sim 0.413
2. Red Orchestra 2: Heroes of Stalingrad (Tripwire Interactive) — 0.409
3. Turok 2: Seeds of Evil (Nightdive Studios) — 0.408
4. GameGuru (The Game Creators) — 0.407
5. Call of Duty®: Modern Warfare® 2 (Infinity Ward) — 0.400

The retrieved games share key characteristics with Valve's portfolio:
- **Tactical/competitive multiplayer** (Red Orchestra 2, CoD MW2 ↔ Counter-Strike, TF2)
- **First-person shooter mechanics** (Turok 2, ROM: Extraction ↔ Half-Life series)
- **Game creation tools** (GameGuru ↔ Source Filmmaker, Alien Swarm SDK)

The approach successfully identifies games in the **same genre neighborhood** as Valve's titles using only textual descriptions. This demonstrates that Valve has a recognisable descriptive language pattern (tactical combat, multiplayer features, environmental storytelling) that can be detected without developer metadata.

**Limitation:** Similarity is driven by genre vocabulary, not company-specific authorship style. The method identifies Valve-like *content*, not Valve-like *writing style*.

### Question 3.3 (Instacart)
The inventory replenishment team is comparing supply chain cycles for fresh foods versus beverages. To verify turnaround schedules, compare the mean and median delay (in days) between orders for users who purchase from the "produce" department vs. users who purchase from the "alcohol" department.

````python
import pandas as pd
import numpy as np
import kagglehub, os

instacart_path = kagglehub.dataset_download("psparks/instacart-market-basket-analysis")
df_prior       = pd.read_csv(os.path.join(instacart_path, "order_products__prior.csv"))
df_orders      = pd.read_csv(os.path.join(instacart_path, "orders.csv"))
df_products    = pd.read_csv(os.path.join(instacart_path, "products.csv"))
df_departments = pd.read_csv(os.path.join(instacart_path, "departments.csv"))

# Identify produce and alcohol department IDs
produce_id = df_departments[df_departments['department']=='produce']['department_id'].values[0]
alcohol_id = df_departments[df_departments['department']=='alcohol']['department_id'].values[0]
print(f"produce dept_id={produce_id}, alcohol dept_id={alcohol_id}")

# Attach department to prior products; then attach user_id
prior_dept = df_prior.merge(df_products[['product_id','department_id']], on='product_id')
prior_user = prior_dept.merge(df_orders[['order_id','user_id']], on='order_id')

# Users who have purchased from each department (non-exclusive)
produce_users = set(prior_user[prior_user['department_id']==produce_id]['user_id'])
alcohol_users = set(prior_user[prior_user['department_id']==alcohol_id]['user_id'])
print(f"Produce users: {len(produce_users):,}  |  Alcohol users: {len(alcohol_users):,}")

# Orders for each user group (excluding first orders with NaN delay)
all_orders_with_delay = df_orders[df_orders['days_since_prior_order'].notna()]
produce_orders = all_orders_with_delay[all_orders_with_delay['user_id'].isin(produce_users)]
alcohol_orders = all_orders_with_delay[all_orders_with_delay['user_id'].isin(alcohol_users)]

print(f"\n{'Metric':<30} {'Produce':>12} {'Alcohol':>12}")
print("-" * 55)
print(f"{'Mean delay (days)':<30} {produce_orders['days_since_prior_order'].mean():>12.4f} {alcohol_orders['days_since_prior_order'].mean():>12.4f}")
print(f"{'Median delay (days)':<30} {produce_orders['days_since_prior_order'].median():>12.1f} {alcohol_orders['days_since_prior_order'].median():>12.1f}")
print(f"{'Std dev':<30} {produce_orders['days_since_prior_order'].std():>12.4f} {alcohol_orders['days_since_prior_order'].std():>12.4f}")
print(f"{'# orders measured':<30} {len(produce_orders):>12,} {len(alcohol_orders):>12,}")
````

### Conclusion – Q3.3

**The mean and median inter-order delay are nearly identical for produce and alcohol users.**

| Metric | Produce Users | Alcohol Users |
|--------|--------------|---------------|
| Mean delay (days) | 10.97 | 10.77 |
| Median delay (days) | 7.0 | 7.0 |
| # Orders measured | ~2.7M | ~206K |

- The **median is identical at 7 days** for both groups — suggesting a dominant weekly shopping rhythm across both segments.
- The mean difference is only **~0.20 days** — negligible and within measurement noise.
- Contrary to the expectation of a faster alcohol supply cycle (which might apply to dedicated beverage retailers), Instacart users buying alcohol and those buying produce follow essentially the same replenishment cadence.

**Recommendation for logistics:** A uniform weekly replenishment assumption works equally well for both fresh-food and alcohol categories on Instacart. There is no evidence of separate supply chain cycles requiring differentiated warehouse scheduling.

### Question 3.4 (Steam)
The customer support audit team is assessing the impact of direct developer contact channels on user satisfaction. To measure if accessible help channels correlate with better reviews, do games that provide a support email address have higher average user ratings than games that do not?

````python
import pandas as pd
import numpy as np
import kagglehub, os

steam_path = kagglehub.dataset_download("nikdavis/steam-store-games")
df_steam   = pd.read_csv(os.path.join(steam_path, "steam.csv"))
df_support = pd.read_csv(os.path.join(steam_path, "steam_support_info.csv"))

# Positive review ratio (keep only games with at least 1 rating)
total = df_steam['positive_ratings'] + df_steam['negative_ratings']
df_s = df_steam[total > 0].copy()
df_s['pos_ratio'] = df_s['positive_ratings'] / total

# Merge with support info; flag presence of email
df_merged = df_s.merge(
    df_support[['steam_appid','support_email']],
    left_on='appid', right_on='steam_appid', how='left'
)
df_merged['has_email'] = df_merged['support_email'].notna()

# Compare ratings by email presence
summary = df_merged.groupby('has_email')['pos_ratio'].agg(
    mean_pos_ratio='mean',
    median_pos_ratio='median',
    count='count'
).reset_index()
summary['has_email'] = summary['has_email'].map({True:'Has support email', False:'No support email'})
print(summary.to_string(index=False))

has = df_merged[df_merged['has_email']==True]['pos_ratio']
no  = df_merged[df_merged['has_email']==False]['pos_ratio']
print(f"\nDifference (has_email − no_email): {has.mean()-no.mean():.4f}")
print(f"Email coverage: {df_support['support_email'].notna().sum():,} / {len(df_support):,} games "
      f"({df_support['support_email'].notna().mean()*100:.1f}%)")
````

### Conclusion – Q3.4

**Games with a support email address do NOT have higher user ratings — in fact, the opposite is observed.**

| Group | Mean Pos. Ratio | Median Pos. Ratio | Count |
|-------|----------------|------------------|-------|
| Has support email | 71.1% | 75.5% | 23,272 |
| No support email | 73.6% | 77.7% | 3,803 |

- Games **without** a support email have a **+2.5 percentage point higher** mean positive-review ratio.
- The hypothesis that "accessible help channels correlate with better reviews" is **not supported**.

**Possible explanations:**
1. **Selection bias:** Larger, more established publishers (who are more likely to list support emails) also publish more games in competitive genres where reviews are harder to maintain at high ratios.
2. **Causality direction:** Games with quality issues may be *more likely* to add a support email in response to user complaints — creating a spurious negative association.
3. **Small indie games** (no dedicated support infrastructure) may receive warmer niche-community reviews because their user base is a self-selected group of fans.

**Recommendation for customer support audit:** Support channel availability alone is not sufficient to predict or improve review scores. Content quality and community management are stronger determinants of user satisfaction.

#### Action Timeline

#### Data & Methodology Choices
- **Why positive ratio rather than raw positive count?** Raw counts are confounded by game popularity; ratio normalises for audience size.
- **Why left-merge?** To retain all rated games; games not in the support table get `has_email = False`.
- **Limitation:** `support_email` was read as `object` dtype (string), meaning any non-null string (including empty strings) counts as "has email." Inspecting sample values confirms these are real email addresses, so the flag is reliable.
- **Alternative considered:** Whether games have a `support_url` instead of (or in addition to) email, as a broader "support channel availability" metric — could be explored in a follow-up.
